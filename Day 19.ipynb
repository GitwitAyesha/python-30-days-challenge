# **Regression**
Regression in machine learning is a type of supervised learning technique used to predict a continuous outcome variable (dependent variable) based on one or more predictor variables (independent variables). The goal of regression is to establish a relationship between the input variables and the output variable, allowing for the prediction of the output when new input data is provided.
# **Correlation**
it is a statistical method which shows linear relationship between two variable(meaning they change at a constant rate)

# **Simple linear regression**
y = a + bx,linear regression.png
a(intercept)
b(slope)

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

# **Data farming**

```python
# read csv (load dataframe)
dataset = pd.read_csv('/content/drive/MyDrive/Datasets/headbrain.csv')
dataset.head()
```

```python
size = dataset['Head Size(cm^3)']
weight = dataset ['Brain Weight(grams)']
print (size)
print(weight)
```

```python
plt.scatter(size,weight)
plt.show()
```

# **Converting dataFrames into NumPy Arrays**
machine learning handle arrays not dataframes

```python
x = np.array(size).reshape(-1,1)
y = np.array(weight).reshape(-1,1)
print(x)
print(y)
```

# **Training the MOdel**
it's being trained on x and y (any 2 major axis of dataset)

```python
model = LinearRegression()
model.fit(x,y)
```

Mean squared error

```python
regression_model_mse = mean_squared_error(x,y)
print("mean_squared_error:\t",math.sqrt(regression_model_mse))    #model loss is mean error
print("r squared value\t\t",model.score(x,y))
```

Visualization

```python
# visualize the dataset with the fit model using matplotlib
plt.scatter(x,y, color = 'orange')
plt.plot(x,model.predict(x),color = 'black')   # y' = model.predict(x)
# y' = estimated prediction variable
plt.title('Linear Regression')
plt.xlabel('Head size')
plt.ylabel('Brain weight')
```

# **Prediction**

```python
# we can manually predict brain weight also by giving an input of brain size
print('Prediction by the model:\t',model.predict(([[2000]])))
```
